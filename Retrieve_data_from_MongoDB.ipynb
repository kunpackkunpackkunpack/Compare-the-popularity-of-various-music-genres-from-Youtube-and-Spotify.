# รวมข้อมูลทั้งสอง DataFrame เข้าด้วยกัน

```python
# Merge the dataframes based on the id and id_spotify columns
merged_df = pd.merge(df_spotify, df_youtube, left_on='id', right_on='id_Spotify')

# View the merged dataframe
merged_df
```

```python
merged_df = merged_df.drop(['id_Spotify', 'Name_y', 'Artist_y'], axis=1)
```

```python
merged_df = merged_df.rename(columns={"Name_x":"Name", "Artist_x":"Artist"})
```
