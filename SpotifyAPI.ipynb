#ดึงข้อมูลปี 2000 - 2003

```python
result_2000_2003 = []
# Retrieving 1000 songs from 2000-2003
for i in range(0, 1000, 20):
    try:
        tracks = sp.search(q='year:2000-2003', type='track', limit=20, offset=i)
        result_2000_2003.append(tracks)
    except spotipy.SpotifyException as e:
        print("Error:", e)
        time.sleep(5) # wait for 5 seconds before retrying
        continue
    time.sleep(np.random.uniform(2, 5))
```

```python
round = 1
trackInfoList_2000_2003 = []
for result in result_2000_2003:
    for track in result['tracks']['items']:
        # Check if the track is already in the track_info_list
        if any(t['Name'] == track['name'] for t in trackInfoList_2000_2003):
          continue

        # Retrieve track information using sp.track method
        track_info = sp.track(track['id'])

        # Extract relevant information from track_info
        track_info_dict = {}
        track_info_dict['Name'] = track_info['name']
        track_info_dict['AlbumName'] = track_info['album']['name']
        track_info_dict['Artist'] = ','.join([artist['name'] for artist in track_info['artists']])
        track_info_dict['ReleaseDate'] = track_info['album']['release_date']
        track_info_dict['PopularityScore'] = track_info['popularity']
        track_info_dict['Duration'] = track_info['duration_ms']

        # Retrieve genre information for each artist associated with the track
        genres = []
        for artist in track_info['artists']:
            artist_info = sp.artist(artist['id'])
            genres.extend(artist_info['genres'])
        track_info_dict['Genre'] = ','.join(genres)

        # Retrieve audio features using sp.audio_features method
        try:
            audio_features = sp.audio_features(track['id'])[0]
            track_info_dict['Danceability'] = audio_features['danceability']
            track_info_dict['Acousticness'] = audio_features['acousticness']
            track_info_dict['Energy'] = audio_features['energy']
            track_info_dict['Instrumentalness'] = audio_features['instrumentalness']
            track_info_dict['Liveness'] = audio_features['liveness']
            track_info_dict['Loudness'] = audio_features['loudness']
            track_info_dict['Speechiness'] = audio_features['speechiness']
            track_info_dict['Tempo'] = audio_features['tempo']
            track_info_dict['Valence'] = audio_features['valence']
            track_info_dict['TimeSignature'] = audio_features['time_signature']
        except:
            track_info_dict['Danceability'] = None
            track_info_dict['Acousticness'] = None
            track_info_dict['Energy'] = None
            track_info_dict['Instrumentalness'] = None
            track_info_dict['Liveness'] = None
            track_info_dict['Loudness'] = None
            track_info_dict['Speechiness'] = None
            track_info_dict['Tempo'] = None
            track_info_dict['Valence'] = None
            track_info_dict['TimeSignature'] = None

        trackInfoList_2000_2003.append(track_info_dict)
        print("Round :",round)
        round+=1

    time.sleep(np.random.uniform(2, 7))

spotify_a = pd.DataFrame(trackInfoList_2000_2003)
```
