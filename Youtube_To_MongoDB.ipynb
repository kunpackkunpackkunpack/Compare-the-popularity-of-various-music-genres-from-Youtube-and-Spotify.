```python
def check_date(publishedAt):
    abnormal_date_pattern = r'\b(?:Premiered|Started streaming on)\s(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\s\d{1,2},\s\d{4}\b|\b\d+\s\w+\sago\b'
    if re.match(abnormal_date_pattern, publishedAt):
        return publishedAt

df_checkDate = df_youtube["publishedAt"].apply(check_date)
df_test = df_youtube

# Drop the rows that contain abnormal date format
df_filtered = df_test.drop(df_checkDate[df_checkDate.isna()].index)

# Reset the index of the filtered DataFrame
df_filtered = df_filtered.reset_index(drop=True)

df_filtered.sample(20)
```

```python
def extract_date(publishedAt):
    # Define regex patterns for normal date format and abnormal date format
    normal_date_pattern = r'\b(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\s\d{1,2},\s\d{4}\b'
    abnormal_date_pattern = r'\b(?:Premiered|Started streaming on)\s(?:Jan|Feb|Mar|Apr|May|Jun|Jul|\
                              Aug|Sep|Oct|Nov|Dec)\s\d{1,2},\s\d{4}\b|\b\d+\s\w+\sago\b'

    # Check if the date string matches with normal or abnormal date pattern
    if re.match(normal_date_pattern, publishedAt):
        date_str = publishedAt
    elif re.match(abnormal_date_pattern, publishedAt):
        # Extract the date information from abnormal date format
        pattern = r"([A-Z][a-z]+ \d{1,2}, \d{4})"
        match = re.search(pattern, publishedAt)
        if match:
            date_str = match.group(1)
        else:
            date_str = datetime.now().strftime("%b %d, %Y")
    else:
        date_str = datetime.now().strftime("%b %d, %Y")

    # Convert the date string to a datetime object and format it as desired
    date_obj = datetime.strptime(date_str, "%b %d, %Y")
    return datetime.strftime(date_obj, "%Y-%m-%d")

df_youtube["publishedAt"] = df_youtube["publishedAt"].apply(extract_date)
```
