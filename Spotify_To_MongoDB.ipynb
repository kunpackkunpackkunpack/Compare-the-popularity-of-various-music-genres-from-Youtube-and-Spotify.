```python
df_spotify = pd.read_excel('SpotifyDataset.xlsx')
df_spotify
```

```python
#แทนค่าว่างด้วยคำว่า "Unknown"
df_spotify["Genre"].fillna("Unknown", inplace=True)
```

```python
df_spotify[df_spotify["Genre"]=="Unknown"]
```

```python
df_spotify[df_spotify['Danceability'].isnull()]
```

```python
#แทนค่าว่างด้วยค่าเฉลี่ยของแต่ละคอลัมน์
df_spotify.fillna(df_spotify.mean(), inplace=True)
```

```python
df_spotify[df_spotify['Name']=='Falling Asleep']
```

```python
df_spotify.info()
```

```python
df_spotify.isnull().sum()
```

```python
#แปลงข้อมูลวันที่เป็นรูปแบบ Datetime
df_spotify["ReleaseDate"] = pd.to_datetime(df_spotify["ReleaseDate"], format="%Y-%m-%d")
```

```python
df_spotify.info()
```

```python
df_spotify = df_spotify.reset_index()
df_spotify = df_spotify.rename(columns={'index': 'id'})
df_spotify
```
